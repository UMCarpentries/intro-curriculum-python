```python
import seaborn.objects as so

so.Plot(data=gapminder_1997).save("../fig/01-plotting-blank.png", dpi=300)
```

```python
(so.Plot(data=gapminder_1997, x="gdpPercap")
 .save("../fig/01-plotting-x-only.png", dpi=300)
)
```

```python
(so.Plot(gapminder_1997, x="gdpPercap", y="lifeExp")
 .save("../fig/01-plotting-x-y.png", dpi=300)
)
```

```python
(so.Plot(gapminder_1997, x="gdpPercap", y="lifeExp")
 .add(so.Dot())
 .save("../fig/01-plotting-x-y-dot.png", dpi=300)
)
```

```python
(so.Plot(gapminder_1997, x="gdpPercap", y="lifeExp")
 .add(so.Dot())
 .label(x="GDP Per Capita", 
        y="Life Expectancy")
 .save("../fig/01-plotting-x-y-dot-xylabels.png", dpi=300)
)
```

```python
(so.Plot(gapminder_1997, x="gdpPercap", y="lifeExp")
 .add(so.Dot())
 .label(x="GDP Per Capita", 
         y="Life Expectancy", 
         title = "Do people in wealthy countries live longer?")
 .save("../fig/01-plotting-x-y-dot-xylabels-title.png", dpi=300)
)
```

```python
(so.Plot(gapminder_1997, x="gdpPercap", y="lifeExp")
 .add(so.Dot(edgecolor="white"))
 .label(x="GDP Per Capita", 
         y="Life Expectancy", 
         title = "Do people in wealthy countries live longer?")
 .save("../fig/01-plotting-x-y-dot-xylabels-title-edgecolor.png", dpi=300)
)
```

```python
(so.Plot(gapminder_1997, 
         x="gdpPercap", 
         y="lifeExp", 
         color="continent")
 .add(so.Dot(edgecolor="white"))
 .label(x="GDP Per Capita", 
         y="Life Expectancy", 
         title = "Do people in wealthy countries live longer?")
 .save("../fig/01-plotting-x-y-color-dot", bbox_inches='tight', dpi=300)
)
```

```python
(so.Plot(gapminder_1997, 
         x="gdpPercap", 
         y="lifeExp", 
         color="continent")
 .add(so.Dot(edgecolor="white"))
 .label(x="GDP Per Capita", 
         y="Life Expectancy", 
         title = "Do people in wealthy countries live longer?")
 .scale(color="Set1")
 .save("../fig/01-plotting-x-y-color-dot-colorpalette", bbox_inches='tight', dpi=300)
)
```

```python
(so.Plot(gapminder_1997, 
         x="gdpPercap", 
         y="lifeExp", 
         color="continent",
         pointsize="pop")
 .add(so.Dot())
 .label(x="GDP Per Capita", 
         y="Life Expectancy", 
         title = "Do people in wealthy countries live longer?")
 .scale(color="Set1")
 .save("../fig/01-plotting-x-y-color-size-dot", bbox_inches='tight', dpi=300)
)
```
